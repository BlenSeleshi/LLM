```python
import re
def remove_emojis(text):
    emoji_pattern = re.compile(
        "[" 
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251" 
        "]+", 
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)


df['Message'] = df['Message'].apply(remove_emojis)
message_df_clean = df['Message'].apply(remove_emojis)

print(df)
print(message_df_clean)
```

```python
import re

pattern = r'[A-Za-z]+'

df['Message'] = df['Message'].apply(lambda x: re.sub(pattern, '', x))

df['Message'] = df['Message'].apply(lambda x: re.sub(r'[ ]+', ' ', x).strip())

df['Message'] = df['Message'].apply(lambda x: x.replace('\\n', '\n'))

df.to_csv('cleaned_v4.csv', index=False)

df.head()
```

```python
# List of unwanted phrases to be removed
unwanted_phrases = [
    'á‹­áˆ„áŠ•áŠ• ./ á‰°áŒ­áŠá‹Â  á‹«á‹µáˆ­áŒ‰á£ á‰¤á‰°áˆ°á‰¥ á‹­áˆáŠ‘',
    'á‹­áˆ„áŠ•áŠ• ./ á‰°áŒ­áŠá‹Â  á‹«á‹µáˆ­áŒ‰á£ á‰¤á‰°áˆ°á‰¥'
    
]

# Function to remove unwanted phrases while preserving newlines
def remove_unwanted_phrases(message):
    for phrase in unwanted_phrases:
        message = message.replace(phrase, '')  # Remove unwanted phrase
    return message  # Return the message without altering newlines

# Apply the function to the 'Message' column
df['Message'] = df['Message'].apply(remove_unwanted_phrases)

# Save the cleaned DataFrame to a new CSV file with newlines preserved
df.to_csv('cleaned_messages.csv', index=False, lineterminator='\n')

# Display the first few rows to check if the messages are cleaned
df.head(50)
```
